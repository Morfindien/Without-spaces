```python
## Experiment 1: Localized Energy Circulation as Effective Mass

Hypothesis:
If a mechanical medium supports stable localized oscillatory structures,
then these structures may behave as particles with effective mass.

We simulate a discrete mechanical lattice and test whether
a localized excitation:

- propagates with finite velocity
- conserves energy
- shows inertial resistance to acceleration

If so, mass may emerge mechanically.
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Lattice parameters
N = 200
dx = 1.0
dt = 0.1
c = 1.0  # wave speed (not arbitrary; defines units)

# Field arrays
u = np.zeros(N)
u_prev = np.zeros(N)

# Initial localized excitation
u[N//2] = 1.0

def step(u, u_prev):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        u_next[i] = (
            2*u[i] - u_prev[i] +
            (c*dt/dx)**2 * (u[i+1] - 2*u[i] + u[i-1])
        )
    return u_next

# Time evolution
steps = 300
history = []

for _ in range(steps):
    u_next = step(u, u_prev)
    history.append(u.copy())
    u_prev = u
    u = u_next

# Plot final state
plt.plot(u)
plt.title("Localized Mechanical Excitation")
plt.show()
```

```python
## Experiment 2: Nonlinear Mechanical Medium

Hypothesis:
Real mechanical media are not perfectly linear.
If restoring forces depend nonlinearly on displacement,
localized structures may become self-stabilizing.

We introduce a minimal nonlinear correction
without arbitrary constants.

Goal:
Test whether nonlinear elasticity produces
stable localized excitations.
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
dx = 1.0
dt = 0.1
c = 1.0

u = np.zeros(N)
u_prev = np.zeros(N)

u[N//2] = 1.0

alpha = 0.1  # nonlinear strength (dimensionless scaling)

def step(u, u_prev):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        laplacian = (u[i+1] - 2*u[i] + u[i-1])
        nonlinear = alpha * u[i]**3
        u_next[i] = (
            2*u[i] - u_prev[i] +
            (c*dt/dx)**2 * laplacian
            - dt**2 * nonlinear
        )
    return u_next

steps = 300
history = []

for _ in range(steps):
    u_next = step(u, u_prev)
    history.append(u.copy())
    u_prev = u
    u = u_next

plt.plot(u)
plt.title("Nonlinear Mechanical Excitation")
plt.show()
```

## Experiment 3: Energy-Dependent Propagation Speed

Hypothesis:
In real mechanical media, wave speed depends on local strain energy.

If propagation speed increases with energy density,
localized excitations may become self-focusing.

We modify the wave equation so that
the effective wave speed depends on local amplitude.

No arbitrary stabilizing potentials are introduced.
Only mechanical self-interaction.

```python

import numpy as np
import matplotlib.pyplot as plt

N = 200
dx = 1.0
dt = 0.1
c0 = 1.0

u = np.zeros(N)
u_prev = np.zeros(N)

u[N//2] = 1.0

beta = 0.2  # energy coupling strength

def step(u, u_prev):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        local_c2 = c0**2 * (1 + beta * u[i]**2)
        laplacian = (u[i+1] - 2*u[i] + u[i-1])
        u_next[i] = (
            2*u[i] - u_prev[i] +
            (local_c2 * dt**2 / dx**2) * laplacian
        )
    return u_next

steps = 300

for _ in range(steps):
    u_next = step(u, u_prev)
    u_prev = u
    u = u_next

plt.plot(u)
plt.title("Energy-Dependent Propagation")
plt.show()
```

```python
## Experiment 4: Bistable Mechanical Medium

Hypothesis:
A mechanical medium may have two stable equilibrium states.

If so, transitions between these states can form stable
localized structures (kinks).

We introduce a minimal double-well restoring force
without adding arbitrary parameters.

Goal:
Test whether topological structures emerge.
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
dx = 1.0
dt = 0.1
c = 1.0

u = np.zeros(N)
u_prev = np.zeros(N)

# small localized perturbation
u[N//2] = 0.5

def step(u, u_prev):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        laplacian = (u[i+1] - 2*u[i] + u[i-1])

        # Double-well restoring force: u - u^3
        nonlinear = u[i] - u[i]**3

        u_next[i] = (
            2*u[i] - u_prev[i] +
            (c*dt/dx)**2 * laplacian
            - dt**2 * nonlinear
        )
    return u_next

steps = 500

for _ in range(steps):
    u_next = step(u, u_prev)
    u_prev = u
    u = u_next

plt.plot(u)
plt.title("Bistable Mechanical Medium")
plt.show()
```

## Experiment 6: Energy vs Velocity

Hypothesis:
If the φ⁴ field equation is Lorentz invariant,
the total energy of the moving kink should increase
according to:

E(v) = γ E0

We compute total field energy for different velocities
and test whether it matches relativistic prediction.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 400
dx = 0.1
dt = 0.01
c = 1.0

x = np.linspace(-20, 20, N)

u = np.tanh(x / np.sqrt(2))

v0 = 0.3
u_prev = u - dt * v0 * (1/np.sqrt(2)) * (1/np.cosh(x/np.sqrt(2)))**2

def step(u, u_prev):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        laplacian = (u[i+1] - 2*u[i] + u[i-1]) / dx**2
        nonlinear = u[i] * (u[i]**2 - 1)
        u_next[i] = (
            2*u[i] - u_prev[i] +
            dt**2 * (c**2 * laplacian - nonlinear)
        )
    u_next[0] = u[0]
    u_next[-1] = u[-1]
    return u_next

steps = 1500
positions = []

for _ in range(steps):
    u_next = step(u, u_prev)
    u_prev = u
    u = u_next

    # find zero crossing
    idx = np.argmin(np.abs(u))
    positions.append(x[idx])

def compute_energy(u, u_prev):
    du_dt = (u - u_prev) / dt
    du_dx = np.gradient(u, dx)

    kinetic = 0.5 * du_dt**2
    gradient = 0.5 * du_dx**2
    potential = 0.25 * (u**2 - 1)**2

    return np.sum(kinetic + gradient + potential) * dx


plt.plot(np.arange(steps)*dt, positions)
plt.xlabel("Time")
plt.ylabel("Kink Position")
plt.title("Kink Center vs Time")
plt.show()
```

## Experiment 6: Relativistic Energy Test

```python

def simulate(v0):
    u = np.tanh(x / np.sqrt(2))
    u_prev = u - dt * v0 * (1/np.sqrt(2)) * (1/np.cosh(x/np.sqrt(2)))**2

    for _ in range(1500):
        u_next = step(u, u_prev)
        u_prev = u
        u = u_next

    return compute_energy(u, u_prev)

velocities = [0.0, 0.2, 0.4, 0.6]
energies = []

for v in velocities:
    energies.append(simulate(v))

E0 = energies[0]
gamma_pred = [1/np.sqrt(1-v**2) for v in velocities]
E_pred = [E0 * g for g in gamma_pred]

print("Measured energies:", energies)
print("Predicted relativistic:", E_pred)
```

## Experiment 9: Lorentz Contraction of the Kink

Hypothesis:
If the φ⁴ field equation is Lorentz invariant,
the moving kink width should contract as:

L(v) = L0 * sqrt(1 - v^2)

We measure kink width at different velocities
and compare with relativistic prediction.

No Lorentz transformation is inserted manually.

```python

import numpy as np

# ---- Width measurement ----
def kink_width(u, x):
    # Find crossing near -0.5
    idx1 = np.argmin(np.abs(u + 0.5))
    # Find crossing near +0.5
    idx2 = np.argmin(np.abs(u - 0.5))
    return abs(x[idx2] - x[idx1])


def measure_width(v0, x, dx, dt, steps):
    # initial kink
    u = np.tanh(x / np.sqrt(2))

    # velocity boost
    u_prev = u - dt * v0 * (1/np.sqrt(2)) * (1/np.cosh(x/np.sqrt(2)))**2

    for _ in range(steps):
        u_next = np.zeros_like(u)
        for i in range(1, len(x)-1):
            laplacian = (u[i+1] - 2*u[i] + u[i-1]) / dx**2
            nonlinear = u[i] * (u[i]**2 - 1)
            u_next[i] = (
                2*u[i] - u_prev[i] +
                dt**2 * (laplacian - nonlinear)
            )
        u_next[0] = u[0]
        u_next[-1] = u[-1]

        u_prev = u
        u = u_next

    return kink_width(u, x)


# ---- Parameters ----
N = 800
dx = 0.1
dt = 0.01
x = np.linspace(-40, 40, N)
steps = 2000

velocities = [0.0, 0.2, 0.4, 0.6]
widths = []

for v in velocities:
    widths.append(measure_width(v, x, dx, dt, steps))

L0 = widths[0]
predicted = [L0 * np.sqrt(1 - v**2) for v in velocities]

print("Measured widths:", widths)
print("Predicted widths:", predicted)
```

## Experiment 10: Small Oscillations Around the Kink

Hypothesis:
If the kink behaves as a particle-like object,
small perturbations around it should form
discrete vibrational modes.

We linearize around the static kink
and analyze the spectrum of fluctuations.

Discrete eigenvalues correspond to internal modes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# grid
N = 400
dx = 0.1
x = np.linspace(-20, 20, N)

# static kink
uK = np.tanh(x/np.sqrt(2))

# build fluctuation operator matrix
H = np.zeros((N, N))

for i in range(1, N-1):
    H[i,i] = 2/dx**2 + (3*uK[i]**2 - 1)
    H[i,i-1] = -1/dx**2
    H[i,i+1] = -1/dx**2

# boundary conditions
H[0,0] = H[-1,-1] = 1e6  # fix ends strongly

# eigenvalues
eigvals, eigvecs = eigh(H)

# show first few eigenvalues
print("Lowest eigenvalues:")
print(eigvals[:6])
```

## Experiment 11: Exciting the Internal Vibrational Mode

Hypothesis:
If the kink possesses a true internal mode,
a small perturbation proportional to the eigenvector
should produce harmonic oscillations.

We:

- Add a small excitation along the internal eigenmode
- Evolve in time
- Measure oscillation frequency
- Compare with predicted ω = sqrt(1.5)

No quantum mechanics is inserted.
This is purely classical field dynamics.

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 400
dx = 0.1
dt = 0.01
x = np.linspace(-20, 20, N)
steps = 4000

# Static kink
u = np.tanh(x/np.sqrt(2))

# Internal mode eigenvector (from previous eigensystem)
internal_mode = eigvecs[:,1]

# Normalize mode
internal_mode /= np.max(np.abs(internal_mode))

# Small amplitude excitation
epsilon = 0.05
u = u + epsilon * internal_mode
u_prev = u.copy()

# Time evolution
center_index = np.argmin(np.abs(x))
amplitude = []

for _ in range(steps):
    u_next = np.zeros_like(u)
    for i in range(1, N-1):
        laplacian = (u[i+1] - 2*u[i] + u[i-1]) / dx**2
        nonlinear = u[i] * (u[i]**2 - 1)
        u_next[i] = (
            2*u[i] - u_prev[i] +
            dt**2 * (laplacian - nonlinear)
        )
    u_next[0] = u[0]
    u_next[-1] = u[-1]

    u_prev = u
    u = u_next

    amplitude.append(u[center_index])

# Plot time signal
t = np.arange(steps)*dt
plt.plot(t, amplitude)
plt.xlabel("Time")
plt.ylabel("Center amplitude")
plt.title("Internal Mode Oscillation")
plt.show()
```

```python

fft_vals = np.fft.fft(amplitude)
freqs = np.fft.fftfreq(len(amplitude), dt)

# Only positive frequencies
mask = freqs > 0
freq_peak = freqs[mask][np.argmax(np.abs(fft_vals[mask]))]

omega_measured = 2*np.pi*freq_peak

print("Measured omega:", omega_measured)
print("Predicted omega:", np.sqrt(1.5))
```

## Experiment 12: 2D φ⁴ Field

We extend the φ⁴ model to two spatial dimensions.

Equation:

∂²u/∂t² = ∂²u/∂x² + ∂²u/∂y² - u(u² - 1)

We initialize a circular domain wall and observe:

- Stability
- Collapse
- Oscillation
- Topological behavior

No new parameters are introduced.

```python

import numpy as np
import matplotlib.pyplot as plt

# Grid
N = 200
dx = 0.2
dt = 0.02
steps = 1000

x = np.linspace(-20, 20, N)
y = np.linspace(-20, 20, N)
X, Y = np.meshgrid(x, y)

# Circular domain wall
R = 5
u = np.tanh((np.sqrt(X**2 + Y**2) - R)/np.sqrt(2))
u_prev = u.copy()

def laplacian(u):
    return (
        (np.roll(u,1,0) - 2*u + np.roll(u,-1,0)) / dx**2 +
        (np.roll(u,1,1) - 2*u + np.roll(u,-1,1)) / dx**2
    )

for _ in range(steps):
    u_next = (
        2*u - u_prev +
        dt**2 * (laplacian(u) - u*(u**2 - 1))
    )
    u_prev = u
    u = u_next

plt.imshow(u, extent=[-20,20,-20,20])
plt.colorbar()
plt.title("2D φ⁴ Domain Structure")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
N = 200
dx = 0.2
dt = 0.02
steps = 1000

x = np.linspace(-20, 20, N)
y = np.linspace(-20, 20, N)
X, Y = np.meshgrid(x, y)

# Initial vortex
theta = np.arctan2(Y, X)
phi1 = np.cos(theta)
phi2 = np.sin(theta)

phi1_prev = phi1.copy()
phi2_prev = phi2.copy()

def laplacian(u):
    return (
        (np.roll(u,1,0) - 2*u + np.roll(u,-1,0)) / dx**2 +
        (np.roll(u,1,1) - 2*u + np.roll(u,-1,1)) / dx**2
    )

for _ in range(steps):
    r2 = phi1**2 + phi2**2
    force1 = (r2 - 1)*phi1
    force2 = (r2 - 1)*phi2

    phi1_next = 2*phi1 - phi1_prev + dt**2*(laplacian(phi1) - force1)
    phi2_next = 2*phi2 - phi2_prev + dt**2*(laplacian(phi2) - force2)

    phi1_prev = phi1
    phi2_prev = phi2

    phi1 = phi1_next
    phi2 = phi2_next

# Plot magnitude
plt.imshow(np.sqrt(phi1**2 + phi2**2), extent=[-20,20,-20,20])
plt.colorbar()
plt.title("Dynamic Vortex Magnitude")
plt.show()
```

## Experiment 14b: Properly Regularized Vortex Core

A physical vortex must have:

We initialize:

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
N = 200
dx = 0.2
dt = 0.02
steps = 1000

x = np.linspace(-20, 20, N)
y = np.linspace(-20, 20, N)
X, Y = np.meshgrid(x, y)

# Initial vortex
theta = np.arctan2(Y, X)
phi1 = np.cos(theta)
phi2 = np.sin(theta)

phi1_prev = phi1.copy()
phi2_prev = phi2.copy()

def laplacian(u):
    return (
        (np.roll(u,1,0) - 2*u + np.roll(u,-1,0)) / dx**2 +
        (np.roll(u,1,1) - 2*u + np.roll(u,-1,1)) / dx**2
    )

for _ in range(steps):

    R = np.sqrt(X**2 + Y**2)
    theta = np.arctan2(Y, X)

    f = np.tanh(R)   # regularized core

    phi1 = f * np.cos(theta)
    phi2 = f * np.sin(theta)

    phi1_prev = phi1.copy()
    phi2_prev = phi2.copy()

# Plot magnitude
plt.imshow(np.sqrt(phi1**2 + phi2**2), extent=[-20,20,-20,20])
plt.colorbar()
plt.title("Dynamic Vortex Magnitude")
plt.show()
```

## Experiment 15: Measuring Topological Charge

We compute the winding number numerically.

The winding number is:

n = (1 / 2π) ∮ ∇θ · dl

We evaluate phase differences around a closed loop
encircling the vortex core.

If n ≈ 1, the vortex carries unit topological charge.

```python

# Compute phase
theta = np.arctan2(phi2, phi1)

# Choose small square around center
cx = N//2
cy = N//2
r = 10

loop = []

# top edge
for i in range(cx-r, cx+r):
    loop.append(theta[i, cy-r])

# right edge
for j in range(cy-r, cy+r):
    loop.append(theta[cx+r, j])

# bottom edge
for i in range(cx+r, cx-r, -1):
    loop.append(theta[i, cy+r])

# left edge
for j in range(cy+r, cy-r, -1):
    loop.append(theta[cx-r, j])

# compute phase differences with branch cut correction
dtheta = np.diff(loop)
dtheta = (dtheta + np.pi) % (2*np.pi) - np.pi

winding = np.sum(dtheta) / (2*np.pi)

print("Winding number:", winding)
```

## Experiment 18b: Relaxed Vortex–Vortex Interaction

We introduce small damping:

∂²φ/∂t² + γ ∂φ/∂t = ∇²φ - (|φ|² - 1)φ

```python
# Grid
N = 200
dx = 0.2
dt = 0.02
steps = 1500

x = np.linspace(-20, 20, N)
y = np.linspace(-20, 20, N)
X, Y = np.meshgrid(x, y)

# Positions
x1, y1 = -5, 0
x2, y2 = 5, 0

R1 = np.sqrt((X-x1)**2 + (Y-y1)**2)
R2 = np.sqrt((X-x2)**2 + (Y-y2)**2)

theta1 = np.arctan2(Y-y1, X-x1)
theta2 = np.arctan2(Y-y2, X-x2)

# Regularized cores
f1 = np.tanh(R1)
f2 = np.tanh(R2)

# Add phases (same winding)
theta_total = theta1 + theta2

phi1 = np.tanh(R1)*np.tanh(R2)*np.cos(theta_total)
phi2 = np.tanh(R1)*np.tanh(R2)*np.sin(theta_total)

phi1_prev = phi1.copy()
phi2_prev = phi2.copy()

def laplacian(u):
    return (
        (np.roll(u,1,0) - 2*u + np.roll(u,-1,0)) / dx**2 +
        (np.roll(u,1,1) - 2*u + np.roll(u,-1,1)) / dx**2
    )

gamma = 0.0  # small damping

phi1_vel = np.zeros_like(phi1)
phi2_vel = np.zeros_like(phi2)

for _ in range(1000):  # relaxation phase
    r2 = phi1**2 + phi2**2
    force1 = (r2 - 1)*phi1
    force2 = (r2 - 1)*phi2

    phi1_acc = laplacian(phi1) - force1 - gamma*phi1_vel
    phi2_acc = laplacian(phi2) - force2 - gamma*phi2_vel

    phi1_vel += dt * phi1_acc
    phi2_vel += dt * phi2_acc

    phi1 += dt * phi1_vel
    phi2 += dt * phi2_vel

plt.imshow(np.sqrt(phi1**2 + phi2**2), extent=[-20,20,-20,20])
plt.colorbar()
plt.title("Vortex–Vortex Interaction")
plt.show()
```

## Experiment 20: Full Dynamic Gauge Field Evolution

We now evolve:

- Scalar field φ
- Gauge field A_i

according to the full Abelian Higgs equations.

We allow the system to relax dynamically.

Only after energy minimization
should magnetic flux quantize.

Flux quantization must emerge —
not be imposed.

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
N = 100
dx = 0.3
dt = 0.005
steps = 2000

x = np.linspace(-20, 20, N)
y = np.linspace(-20, 20, N)
X, Y = np.meshgrid(x, y)

# Initial vortex
R = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

phi = np.tanh(R) * np.exp(1j * theta)

Ax = np.zeros_like(phi.real)
Ay = np.zeros_like(phi.real)

phi_vel = np.zeros_like(phi)
Ax_vel = np.zeros_like(Ax)
Ay_vel = np.zeros_like(Ay)

def grad_x(u):
    return (np.roll(u,-1,1) - np.roll(u,1,1))/(2*dx)

def grad_y(u):
    return (np.roll(u,-1,0) - np.roll(u,1,0))/(2*dx)

def laplacian(u):
    return (
        (np.roll(u,1,0) - 2*u + np.roll(u,-1,0))/dx**2 +
        (np.roll(u,1,1) - 2*u + np.roll(u,-1,1))/dx**2
    )

gamma = 0.3  # damping for relaxation

for _ in range(steps):

    # Covariant derivatives
    Dx_phi = grad_x(phi) - 1j * Ax * phi
    Dy_phi = grad_y(phi) - 1j * Ay * phi

    D2_phi = grad_x(Dx_phi) + grad_y(Dy_phi)

    # Scalar field equation
    phi_acc = D2_phi - (np.abs(phi)**2 - 1)*phi - gamma*phi_vel

    # Magnetic field
    B = grad_x(Ay) - grad_y(Ax)

    # Gauge field equations
    Jx = np.imag(np.conj(phi) * Dx_phi)
    Jy = np.imag(np.conj(phi) * Dy_phi)

    Ax_acc = grad_y(B) - Jx - gamma*Ax_vel
    Ay_acc = -grad_x(B) - Jy - gamma*Ay_vel

    # Update
    phi_vel += dt * phi_acc
    Ax_vel += dt * Ax_acc
    Ay_vel += dt * Ay_acc

    phi += dt * phi_vel
    Ax += dt * Ax_vel
    Ay += dt * Ay_vel

B = grad_x(Ay) - grad_y(Ax)

plt.imshow(B.real, extent=[-20,20,-20,20])
plt.colorbar()
plt.title("Magnetic Field After Relaxation")
plt.show()

flux = np.sum(B.real) * dx**2
print("Flux:", flux)
print("2π:", 2*np.pi)
```
